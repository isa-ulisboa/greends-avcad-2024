```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PowerTransformer
from sklearn.metrics import r2_score
import statsmodels.api as sm

# Load the dataset
# Assuming the dataframe is named df2 and contains the relevant variables

# Define the continuous environmental variables
continuous_vars = ['Altitude', 'Actual_river_slope', 'Elevation_mean_catch',
                   'prec_ann_catch', 'temp_ann', 'temp_jan', 'temp_jul']

# Check if any variable requires transformation and apply the appropriate transformation
vars_to_transform = []
for var in continuous_vars:
    if np.any(df2[var] <= 0):  # Check for non-positive values
        vars_to_transform.append(var)

# Apply power transformation to variables that require transformation
for var in vars_to_transform:
    df2[var + '_transformed'] = PowerTransformer().fit_transform(df2[[var]])

# Run simple linear regression for each variable
results = {}
for var in continuous_vars:
    if var in vars_to_transform:
        X = df2[var + '_transformed'].values.reshape(-1, 1)
    else:
        X = df2[var].values.reshape(-1, 1)
    y = df2['Species_Richness'].values.reshape(-1, 1)

    # Add a constant term to the predictor
    X = sm.add_constant(X)

    # Fit the linear regression model
    model = sm.OLS(y, X)
    results[var] = model.fit()

# Print regression coefficients, R-squared, and F-statistics
for var, result in results.items():
    print(f"Variable: {var}")
    print(f"Regression Coefficients: {result.params}")
    print(f"R-squared: {result.rsquared}")
    print(f"F-statistic: {result.fvalue}")
    print("\n")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances



# Filter sites from the Douro and Tejo basins
douro_tejo_sites = df2[(df2['Catchment_name'] == 'Douro') | (df2['Catchment_name'] == 'Tejo')]

# Select quantitative environmental variables
quantitative_vars = ['Altitude', 'Actual_river_slope', 'Elevation_mean_catch',
                     'prec_ann_catch', 'temp_ann', 'temp_jan', 'temp_jul']

# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(douro_tejo_sites[quantitative_vars])

# Define linkage methods
linkage_methods = ['ward', 'complete', 'average', 'single']

# Perform agglomerative clustering with different linkage methods
for method in linkage_methods:
    # Fit Agglomerative clustering model
    clustering = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage=method)
    clustering.fit(X)

    # Add cluster labels to the DataFrame
    douro_tejo_sites[f'Cluster_{method}'] = clustering.labels_

# Visualize the clusters
plt.figure(figsize=(15, 5))
for i, method in enumerate(linkage_methods, 1):
    plt.subplot(1, len(linkage_methods), i)
    sns.scatterplot(data=douro_tejo_sites, x='Longitude', y='Latitude', hue=f'Cluster_{method}', palette='viridis')
    plt.title(f'Clusters with {method} linkage')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming df2 is your dataframe
# Filter sites from the Douro and Tejo basins
douro_tejo_sites = df2[(df2['Catchment_name'] == 'Douro') | (df2['Catchment_name'] == 'Tejo')]

# Select quantitative environmental variables
quantitative_vars = ['Altitude', 'Actual_river_slope', 'Elevation_mean_catch',
                     'prec_ann_catch', 'temp_ann', 'temp_jan', 'temp_jul']

# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(douro_tejo_sites[quantitative_vars])

# Perform PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['Catchment_name'] = douro_tejo_sites['Catchment_name']

# Plot the biplot
plt.figure(figsize=(10, 7))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Catchment_name', palette='viridis')
plt.title('PCA Biplot')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.2f}% variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.2f}% variance)')

# Adding vectors for the features
for i, var in enumerate(quantitative_vars):
    plt.arrow(0, 0, pca.components_[0, i], pca.components_[1, i],
              color='r', alpha=0.5, head_width=0.05)
    plt.text(pca.components_[0, i] * 1.15, pca.components_[1, i] * 1.15, var,
             color='g', ha='center', va='center')

plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import pdist, squareform
from scipy.linalg import eigh

# Assuming df2 is your dataframe
# Filter sites from the Douro and Tejo basins
douro_tejo_sites = df2[(df2['Catchment_name'] == 'Douro') | (df2['Catchment_name'] == 'Tejo')]

# Select quantitative environmental variables
quantitative_vars = ['Altitude', 'Actual_river_slope', 'Elevation_mean_catch',
                     'prec_ann_catch', 'temp_ann', 'temp_jan', 'temp_jul']

# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(douro_tejo_sites[quantitative_vars])

# Compute the distance matrix
distance_matrix = pdist(X, metric='euclidean')
distance_matrix = squareform(distance_matrix)

# Perform PCoA
n = distance_matrix.shape[0]
H = np.eye(n) - np.ones((n, n)) / n
B = -H.dot(distance_matrix ** 2).dot(H) / 2

# Eigen decomposition
eigenvalues, eigenvectors = eigh(B)
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Create a DataFrame with the first two PCoA axes
pcoa_df = pd.DataFrame(eigenvectors[:, :2], columns=['PCoA1', 'PCoA2'])
pcoa_df['Catchment_name'] = douro_tejo_sites['Catchment_name'].values

# Plot the PCoA results
plt.figure(figsize=(10, 7))
sns.scatterplot(data=pcoa_df, x='PCoA1', y='PCoA2', hue='Catchment_name', palette='viridis')
plt.title('PCoA Projection')
plt.xlabel(f'PCoA1 ({eigenvalues[0]/sum(eigenvalues)*100:.2f}% variance)')
plt.ylabel(f'PCoA2 ({eigenvalues[1]/sum(eigenvalues)*100:.2f}% variance)')
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Assuming df2 is your dataframe
# Filter sites from the Douro and Tejo basins
douro_tejo_sites = df2[(df2['Catchment_name'] == 'Douro') | (df2['Catchment_name'] == 'Tejo')]

# Select quantitative environmental variables
quantitative_vars = ['Altitude', 'Actual_river_slope', 'Elevation_mean_catch',
                     'prec_ann_catch', 'temp_ann', 'temp_jan', 'temp_jul']

# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(douro_tejo_sites[quantitative_vars])
y = douro_tejo_sites['Catchment_name']

# Perform LDA
lda = LDA(n_components=1)  # Only one component since we have two classes
X_lda = lda.fit_transform(X, y)

# Create a DataFrame with the LDA results
lda_df = pd.DataFrame(X_lda, columns=['LD1'])
lda_df['Catchment_name'] = y.values

# Plot the LDA results
plt.figure(figsize=(10, 7))
sns.scatterplot(data=lda_df, x='LD1', y=[0]*len(lda_df), hue='Catchment_name', palette='viridis')
plt.title('LDA Biplot')
plt.xlabel('LD1')
plt.ylabel('')
plt.yticks([])  # Hide y-axis ticks as they are not meaningful here
plt.grid()
plt.show()

# Create biplot
fig, ax = plt.subplots(figsize=(10, 7))

# Scatter plot
scatter = ax.scatter(lda_df['LD1'], np.zeros_like(lda_df['LD1']), c=lda_df['Catchment_name'].map({'Douro': 0, 'Tejo': 1}), cmap='viridis', label=lda_df['Catchment_name'])

# Create a legend
legend1 = ax.legend(*scatter.legend_elements(), title='Catchment_name')
ax.add_artist(legend1)

# Plot the loading vectors
for i, var in enumerate(quantitative_vars):
    plt.arrow(0, 0, lda.coef_[0, i] * 5, 0, color='red', alpha=0.5)
    plt.text(lda.coef_[0, i] * 5.2, 0.2, var, color='black', ha='center', va='center')

plt.title('LDA Biplot with Loading Vectors')
plt.xlabel('LD1')
plt.ylabel('')
plt.yticks([])  # Hide y-axis ticks as they are not meaningful here
plt.grid()
plt.show()

```
